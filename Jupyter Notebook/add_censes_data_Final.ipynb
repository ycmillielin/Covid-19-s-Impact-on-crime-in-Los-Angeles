# clean

```python
unemploy = unemploy.astype({'zipcode':'str'})
income = income.astype({'zipcode':'str'})

pop = pop.loc[:, ['Total population', 'Year', 'zipcode']]
unemploy = unemploy.loc[:, ['Unemployment_rate', 'Year', 'zipcode']]
```

```python
df = pd.merge(pop, unemploy, on = ['Year', 'zipcode'], how = 'left')
df = pd.merge(df, income, on = ['Year', 'zipcode'], how = 'left')
df.head()
```

# Calculate the Unemployment_rate, Median household, income data in year 2010, 2020, 2021

## calculate average percentage of each data in each zip code

```python
# population
pop_by_year = df.groupby('Year')[['Total population']].sum()
df = pd.merge(df, pop_by_year, on = 'Year', how = 'left')
df.head()
```

```python
df = df.rename(columns = {'Total population_y': 'Total_pop_year'})
df = df.rename(columns = {'Total population_x': 'Total population'})
df.head()
```

```python
df['Total_pop_per'] = df['Total population'] / df['Total_pop_year']
```

```python
total_pop_per = df.groupby('zipcode')[['Total_pop_per']].mean()
total_pop_per.head()
```

```python
# unemployment rate
unem_year = df.groupby('Year')[['Unemployment_rate']].mean()
df = pd.merge(df, unem_year, on = 'Year', how = 'left')
df.head()
```

```python
df = df.rename(columns = {'Unemployment_rate_y': 'Unemployment_rate_year'})
df = df.rename(columns = {'Unemployment_rate_x': 'Unemployment_rate'})
df.head()
```

```python
df['unem_per'] = df['Unemployment_rate'] / df['Unemployment_rate_year']
```

```python
unem_per =df.groupby('zipcode')[['unem_per']].mean()
unem_per.head()
```

```python
# income
income_year = df.groupby('Year')[['Median household income']].mean()
df = pd.merge(df, income_year, on = 'Year', how = 'left')
df.head()
```

```python
df = df.rename(columns = {'Median household income_y': 'income_year'})
df = df.rename(columns = {'Median household income_x': 'Median household income'})
df.head()
```

```python
df['income_per'] = df['Median household income'] / df['income_year']
```

```python
income_per = df.groupby('zipcode')[['income_per']].mean()
income_per.head()
```

## caculate the assumed data in year 2010, 2020, 2021

```python
# calculate 2010, 2020, 2021
# population
# 2010: 9818605
# 2020: 10014009
# 2021: 9969510
```

```python
total_pop_per['Year'] = 2021
total_pop_per.head()
```

```python
total_pop_per['Total population'] = round(total_pop_per['Total_pop_per'] * 9969510)
total_pop_per = total_pop_per.reset_index()
total_pop_per.head()
```

```python
# 2021
df21 = total_pop_per.loc[:, ['Year', 'zipcode', 'Total population']]
df21.head()
```

```python
# 2020
total_pop_per['Year'] = 2020
total_pop_per['Total population'] = round(total_pop_per['Total_pop_per'] * 10014009)
df20 = total_pop_per.loc[:, ['Year', 'zipcode', 'Total population']]
df20.head()
```

```python
# 2010
total_pop_per['Year'] = 2010
total_pop_per['Total population'] = round(total_pop_per['Total_pop_per'] * 9818605)
df10 = total_pop_per.loc[:, ['Year', 'zipcode', 'Total population']]
df10.head()
```

```python
# Unemployment Rate
# 2010: 12.6
# 2020: 12.9
# 2021: 11.1
```

```python
# 2021
unem_per['Year'] = 2021
unem_per['Unemployment_rate'] = unem_per['unem_per'] * 11.1
unem_per = unem_per.reset_index()
df21 = pd.merge(df21, unem_per, on = 'zipcode', how = 'left')
df21.head()
```

```python
df21 = df21.drop(columns=['Year_y', 'unem_per'])
df21 = df21.rename(columns = {'Year_x': 'Year'})
df21.head()
```

```python
# 2020
unem_per['Year'] = 2020
unem_per['Unemployment_rate'] = unem_per['unem_per'] * 12.9
df20 = pd.merge(df20, unem_per, on = 'zipcode')
df20 = df20.drop(columns=['Year_y', 'unem_per'])
df20 = df20.rename(columns = {'Year_x': 'Year'})
df20.head()
```

```python
# 2010
unem_per['Year'] = 2010
unem_per['Unemployment_rate'] = unem_per['unem_per'] * 12.6
df10 = pd.merge(df10, unem_per, on = 'zipcode')
df10 = df10.drop(columns=['Year_y', 'unem_per'])
df10 = df10.rename(columns = {'Year_x': 'Year'})
df10.head()
```

```python
# Median household income
# 2021: 80000
# 2020: 70700
# 2010: 52595
```

```python
# 2021
income_per['Year'] = 2021
income_per['Median household income'] = income_per['income_per'] * 80000
income_per = income_per.reset_index()
df21 = pd.merge(df21, income_per, on = 'zipcode', how = 'left')
df21 = df21.drop(columns=['Year_y'])
df21 = df21.rename(columns = {'Year_x': 'Year'})
df21.head()
```

```python
df21 = df21.drop(columns=['income_per'])
df21.head()
```

```python
# 2020
income_per['Year'] = 2020
income_per['Median household income'] = income_per['income_per'] * 70700
df20 = pd.merge(df20, income_per, on = 'zipcode', how = 'left')
df20 = df20.drop(columns=['Year_y', 'income_per'])
df20 = df20.rename(columns = {'Year_x': 'Year'})
df20.head()
```

```python
# 2010
income_per['Year'] = 2010
income_per['Median household income'] = income_per['income_per'] * 52595
df10 = pd.merge(df10, income_per, on = 'zipcode', how = 'left')
df10 = df10.drop(columns=['Year_y', 'income_per'])
df10 = df10.rename(columns = {'Year_x': 'Year'})
df10.head()
```

```python
# choose useful columns from df
final = df.loc[:, ['Year', 'zipcode', 'Total population', 'Unemployment_rate', 'Median household income']]
final.head()
```
