```python
import pandas as pd
import numpy as np
import pickle
pd.set_option('display.max_columns', None)
```

```python
file = open("dataset_w_Zipcode.sav",'rb')
df = pickle.load(file)
file.close()
```

```python
zip_area = df.groupby(['zipcode','AREA_NAME'])[['DR_NO']].count().reset_index()
zip_area
```

```python
zip_areas = zip_area.loc[zip_area.groupby("zipcode")["DR_NO"].idxmax()]
zip_areas
```

```python
zip_areas.to_csv('zip_area_max.csv', index=False)
```
