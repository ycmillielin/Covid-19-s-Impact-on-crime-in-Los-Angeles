```python
import pandas as pd
import numpy as np
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import requests
import json

# import data
df = pd.read_csv('coords.csv')

my_key = "Your key"

final_zips = {}
for j in df['coords']:
    print(j)
    if j == '0.0,0.0':
        continue
    response = requests.get("https://maps.googleapis.com/maps/api/geocode/json?latlng="+str(j)+"&key="+my_key).text
    response_info = json.loads(response)
    address = response_info['results'][1]['formatted_address']
    try:
        final_zips[j] = address.split(' CA ')[1].split(',')[0]
        print(address.split(' CA ')[1].split(', USA')[0])
    except:
        try:
            final_zips[j] = address.split(' California ')[1].split(',')[0]
            print(address.split(' California ')[1].split(', USA')[0]) 
        except:
            final_zips[j] = 'NOT FOUND'
    if final_zips[j] == 'NOT FOUND':
        try:
            address = response_info['results'][2]['formatted_address']
            final_zips[j] = address.split(' CA ')[1].split(',')[0]
            print(address.split(' CA ')[1].split(', USA')[0])
        except:
            final_zips[j] = 'NOT FOUND'
            print(final_zips[j])
final_df = pd.DataFrame({'coords':final_zips.keys(),'zipcode':final_zips.values()})
final_df.to_csv('Coords_w_Zipcodes.csv',header = True,index = False)
```
